```python
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
def generate_data(w,b,alpha):
    x=np.random.rand(10)*100
    y=w*x + b
    y=np.array([i+alpha*np.random.random() for i in y])
    return (x,y)
```

```python
X,Y = generate_data(0.5,6,15)
sns.pointplot(X,Y)
```

```python
def forward_prop(x,w,b):
    return w*x+b
```

```python
def grad_e_w(w,b,x,y_in):
    return (w*x+b-y_in)*2*w
def grad_e_b(w,b,x,y_in):
    return (w*x+b-y_in)*2*1
def error(w,b,x,y_in):
    return (w*x+b-y_in)**2
```

```python
w=5
b=np.random.rand()
W,B=[],[]
learning_rate = 0.000001
for i in range(1000):
    for x,y in zip(X,Y):
        w_prev=w
        b_prev=w
        w=w_prev-learning_rate*grad_e_w(w_prev,b_prev,x,y)
        b=b_prev-learning_rate*grad_e_b(w_prev,b_prev,x,y)
    W.append(w)
    B.append(b)
    
```
